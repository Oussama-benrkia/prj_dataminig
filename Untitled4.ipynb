```python
import pandas as pd
import numpy as np
from scipy.special import softmax  # Pour convertir les scores en probabilités
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, recall_score, roc_curve, roc_auc_score
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, f1_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df.head()
```

# New Section

```python
target_distribution = df['price_range'].value_counts(normalize=True) * 100
print("price_range variable distribution (%):\n", target_distribution)
```

```python
X, y = df.drop('price_range', axis=1), df['price_range']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train_scaled, y_train)
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)
```

```python
models = {
    "KNN": knn,
    "Logistic Regression": logistic_regression,
    "SVC": svc
}

for name, model in models.items():
    y_pred = model.predict(X_test_scaled if name in ["KNN", "Logistic Regression", "SVC"] else X_test)
    print(f"{name} Classification Report:")
    print(classification_report(y_test, y_pred))
    print("-" * 55)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def evaluate_model_performance(models, X_test, X_test_scaled, y_test):
    for name, model in models.items():
        # Sélection des features test selon le modèle
        X_used = X_test_scaled if name in ["KNN", "Logistic Regression", "SVC"] else X_test
        y_pred = model.predict(X_used)

        print(f"Model: {name}")
        print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
        print(f"Precision: {precision_score(y_test, y_pred, average='macro'):.2f}")
        print(f"Recall: {recall_score(y_test, y_pred, average='macro'):.2f}")
        print(f"F1 Score: {f1_score(y_test, y_pred, average='macro'):.2f}")
        print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
        print("-" * 25)

evaluate_model_performance(models, X_test, X_test_scaled, y_test)
```

```python
from sklearn.metrics import recall_score

def evaluate_recall(model, X_test, y_test):
    y_pred = model.predict(X_test)
    return recall_score(y_test, y_pred, average="macro")  # Correction ici

print("KNN Recall:", evaluate_recall(knn, X_test_scaled, y_test))
print("Logistic Regression Recall:", evaluate_recall(logistic_regression, X_test_scaled, y_test))
print("SVC Recall:", evaluate_recall(svc, X_test_scaled, y_test))
```

```python

plt.figure(figsize=(10, 6))

for name, model in models.items():
    if hasattr(model, "predict_proba"):
        y_pred_prob = model.predict_proba(X_test_scaled if name in ["KNN", "Logistic Regression", "SVC"] else X_test)
    else:
        # Utiliser decision_function et appliquer softmax
        y_scores = model.decision_function(X_test_scaled if name in ["KNN", "Logistic Regression", "SVC"] else X_test)
        y_pred_prob = softmax(y_scores, axis=1)  # Convertir en probas

    # Tracer une courbe ROC pour chaque classe
    for i in range(y_pred_prob.shape[1]):
        fpr, tpr, _ = roc_curve(y_test, y_pred_prob[:, i], pos_label=i)
        auc = roc_auc_score(y_test, y_pred_prob, multi_class="ovr", average="macro")

        plt.plot(fpr, tpr, label=f"{name} (Class {i}, AUC: {auc:.3f})")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.title("ROC Curve Comparison (Multi-Class)")
plt.show()
```

```python


def plot_model_comparison(models, X_test, X_test_scaled, y_test):
    metrics = {"Accuracy": [], "Precision": [], "Recall": [], "F1 Score": []}
    model_names = []

    for name, model in models.items():
        y_pred = model.predict(X_test_scaled if name in ["KNN", "Logistic Regression", "SVC"] else X_test)

        metrics["Accuracy"].append(accuracy_score(y_test, y_pred))
        metrics["Precision"].append(precision_score(y_test, y_pred, average="macro"))  # Correction ici
        metrics["Recall"].append(recall_score(y_test, y_pred, average="macro"))  # Correction ici
        metrics["F1 Score"].append(f1_score(y_test, y_pred, average="macro"))  # Correction ici
        model_names.append(name)

    # Création du DataFrame
    metrics_df = pd.DataFrame(metrics, index=model_names)

    # Affichage du graphe
    metrics_df.plot(kind='bar', figsize=(10, 6))
    plt.title("Model Performance Comparison")
    plt.ylabel("Score")
    plt.xlabel("Models")
    plt.legend(loc="lower right")
    plt.show()

plot_model_comparison(models, X_test, X_test_scaled, y_test)
```
